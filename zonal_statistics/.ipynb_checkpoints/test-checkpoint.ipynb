```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
geom = gpd.read_file('geom/buffer_10km_clipped.gpkg')[['LOCATION_ID','geometry']]
data = pd.read_csv('out/pop_10km_clipped.csv')
merge = pd.merge(geom, data, on='LOCATION_ID', how='inner')
```

```python
fig,ax = plt.subplots()
merge.plot(column='pop_2021', vmin=0, vmax=1000, cmap='Greens', 
           edgecolor='r', ax=ax)
ax.set_axis_off()
plt.tight_layout()
plt.show()
```
